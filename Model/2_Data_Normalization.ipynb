```python
#normalisasi data menggunakan Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# Membaca data dari file CSV
data1 = pd.read_csv('/content/drive/MyDrive/Dataset/new_data.csv')

# Memilih hanya 8 features yang akan dinormalisasi
features_to_normalize = ['pH', 'TSS', 'DO', 'BOD', 'COD', 'Nitrat', 'FecalColiform', 'Fosfat']

# Normalisasi menggunakan Min-Max Scaling
scaler1 = MinMaxScaler()
data1[features_to_normalize] = scaler1.fit_transform(data[features_to_normalize])

# Menampilkan data setelah normalisasi
print("\nSetelah Normalisasi menggunakan Min-Max Scaling:")
print(data1.head())
```
