```python
from imblearn.over_sampling import SMOTE, SMOTEN
from imblearn.combine import SMOTEENN
from sklearn.model_selection import train_test_split

# Assuming 'X' is your feature matrix, and 'y' is your target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

# Print the distribution of classes in the training set before applying SMOTE
print("Distribution of Classes in Training Set (Before SMOTE):")
print(y_train.value_counts())
min_class_samples = min(y_train.value_counts())
# Apply SMOTE
smote = SMOTE(sampling_strategy='auto', k_neighbors=min_class_samples - 1, random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Print the distribution of classes in the training set after applying SMOTE
print("\nDistribution of Classes in Training Set (After SMOTE):")
print(y_train_smote.value_counts())


```

```python
import pandas as pd

# Assuming X_train_smote contains your feature data and y_train_smote contains your target data
# Convert arrays to DataFrame
X_train_smote_df = pd.DataFrame(X_train_smote, columns=['BOD', 'COD', 'FecalColiform', 'IP'])  # Assuming these are your feature names
y_train_smote_df = pd.DataFrame(y_train_smote, columns=['Class'])

# Concatenate X_train_smote_df and y_train_smote_df
data = pd.concat([X_train_smote_df, y_train_smote_df], axis=1)

# Save to CSV
data.to_csv('smote_data0.8.csv', index=False)
```
